```python
# THEORY QUESTIONS WITH EXAMPLES

# 1. What is the difference between a function and a method in Python?
# A function is a block of reusable code, while a method is a function that is associated with an object.
# Example:
def greet():
    return "Hello"
print(greet())

# Method Example:
text = "hello"
print(text.upper())  # upper() is a method of string object

# 2. Function arguments and parameters in Python
# Parameters are variables in the function definition, arguments are values passed to the function.
def add(a, b):  # a and b are parameters
    return a + b
print(add(3, 4))  # 3 and 4 are arguments

# 3. Ways to define and call a function
# Define with def keyword
def say_hello(name):
    return f"Hello, {name}"
print(say_hello("Anubhav"))

# 4. Purpose of return statement
def square(n):
    return n * n
print(square(5))  # Outputs 25

# 5. Iterators vs Iterables
# Iterable: can return an iterator (e.g., list, string). Iterator: has __next__() method.
my_list = [1, 2, 3]
iterator = iter(my_list)
print(next(iterator))  # 1

# 6. Generators in Python
# Generators use yield to produce a sequence of values lazily
def my_generator():
    yield 1
    yield 2
    yield 3
for val in my_generator():
    print(val)

# 7. Advantages of Generators
# - Memory efficient
# - Useful for large datasets
# Example:
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
print(list(count_up_to(5)))

# 8. Lambda function
# Small anonymous function used for short operations
add = lambda x, y: x + y
print(add(2, 3))

# 9. map() function
# Applies a function to all elements in an iterable
nums = [1, 2, 3]
squares = list(map(lambda x: x**2, nums))
print(squares)

# 10. map vs reduce vs filter
from functools import reduce

# map applies a function
print(list(map(lambda x: x*2, [1, 2, 3])))  # [2, 4, 6]

# filter filters values
print(list(filter(lambda x: x % 2 == 0, [1, 2, 3])))  # [2]

# reduce reduces to single value
print(reduce(lambda x, y: x + y, [1, 2, 3]))  # 6

# 11. Internal Mechanism of reduce (Pen & Paper Required)
#  Use reduce(lambda x,y: x+y, [47,11,42,13]) â†’ (((47+11)+42)+13)

# PRACTICAL QUESTIONS

# 1. Sum of even numbers
def sum_even(nums):
    return sum(x for x in nums if x % 2 == 0)
print(sum_even([1, 2, 3, 4, 5, 6]))

# 2. Reverse a string
def reverse_string(s):
    return s[::-1]
print(reverse_string("hello"))

# 3. Squares of list
def square_list(nums):
    return [x**2 for x in nums]
print(square_list([1, 2, 3]))

# 4. Check for prime numbers from 1 to 200
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
primes = [n for n in range(1, 201) if is_prime(n)]
print(primes)

# 5. Fibonacci iterator class
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.a, self.b, self.count = 0, 1, 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

for num in Fibonacci(10):
    print(num)

# 6. Generator for powers of 2
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

print(list(powers_of_two(5)))

# 7. Generator to read file line by line
def read_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()

# Example:
# for line in read_lines("sample.txt"):
#     print(line)

# 8. Sort tuples by second element
data = [(1, 3), (2, 1), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)

# 9. Celsius to Fahrenheit using map
celsius = [0, 10, 20, 30]
fahrenheit = list(map(lambda c: (9/5)*c + 32, celsius))
print(fahrenheit)

# 10. Remove vowels using filter
def remove_vowels(s):
    return ''.join(filter(lambda ch: ch.lower() not in 'aeiou', s))
print(remove_vowels("Hello World"))

# 11. Bookshop accounting logic
orders = [
    ["34587", "Learning Python", 4, 40.95],
    ["98762", "Programming in C", 5, 56.80],
    ["77226", "Head First Java", 3, 32.95],
    ["88112", "Data Science", 3, 24.99]
]

result = list(map(lambda x: (x[0], x[2] * x[3] + 10 if x[2] * x[3] < 100 else x[2] * x[3]), orders))
print(result)
```
